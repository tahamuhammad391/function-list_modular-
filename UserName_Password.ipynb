```python
## Make a List of user!!!
log=[]
def admin():
    while True:
        nu= int(input("Enter a number of user you want to create"))
        for i in range(nu):
            userName = input(f"Enter a user {i+1} name").lower()
            password = input(f"Enter a {i+1} password").lower()
            log.append([userName,password])
        check=input("What you want to check \n1 Count o user \n2 Want to Search any User \n3 Want to remove any user ")
        if check=="1":
            print(f"Total Number of User is = {len(log)}")
            break
        if check=="2":
            print("Search for User Portal")
            lst=[]
            for i in range(len(log)):
                lst.append(log[i][0])
            check2 = input("Enter a Name you want to Search").lower()
            count= lst.count(check2)
            if count>0:
                print("User Exist!!!")
                break
            else:
                print("User Does Not Exist")
                break
        if check=="3":
            print("Remove the User from Portal")
            lst=[]
            for i in range(len(log)):
                lst.append(log[i][0])
            check2 = input("Enter a Name you want to Remove").lower()
            var= lst.index(check2)
            print(log.pop(var))
            print(f"Updated User is :{log}")
            break
            
            
```

```python
admin()
```
